```python
import requests
import os
from pathlib import Path
from dotenv import load_dotenv

env_path = ".env"
load_dotenv(env_path)
NEWS_API_KEY = os.getenv("GNEWS_API_KEY")

print(f"API Key: {NEWS_API_KEY[:10]}...{NEWS_API_KEY[-5:]}")

url = "https://gnews.io/api/v4/search"
params = {
    "q": "Reliance Industries ESG",
    "lang": "en",
    "country": "in",
    "max": 5,
    "apikey": NEWS_API_KEY,
}

resp = requests.get(url, params=params)
print(f"Status: {resp.status_code}")
print(f"Response: {resp.json()}")
```

```python
import feedparser
import requests

company = "Reliance Industries"
query = f"{company} ESG pollution SEBI"

url = f"https://news.google.com/rss/search?q={requests.utils.quote(query)}&hl=en-IN&gl=IN&ceid=IN:en"

print(f"Testing RSS for: {company}")
print(f"URL: {url}\n")

feed = feedparser.parse(url)

print(f"Status: {feed.get('status', 'N/A')}")
print(f"Total entries: {len(feed.entries)}\n")

for idx, entry in enumerate(feed.entries[:5], 1):
    print(f"{idx}. {entry.title}")
    print(f"   Source: {entry.source.title if hasattr(entry, 'source') else 'N/A'}")
    print(f"   Link: {entry.link}")
    print(f"   Published: {entry.published if hasattr(entry, 'published') else 'N/A'}\n")
```

```python
import os
import requests
from pathlib import Path
from dotenv import load_dotenv

# Load .env file from two levels up
env_path = Path.cwd().parents[1] / ".env"
load_dotenv(dotenv_path=env_path)

# Get API key
api_key = os.environ.get("GROQ_API_KEY")
if not api_key:
    raise ValueError("GROQ_API_KEY environment variable is not set")

# Define the API endpoint
url = "https://api.groq.com/openai/v1/models"

# Set headers with authorization
headers = {
    "Authorization": f"Bearer {api_key}",
    "Content-Type": "application/json"
}

# Send GET request to fetch models
response = requests.get(url, headers=headers)

# Print the JSON response
print(response.json())   
```
